```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.cluster import KMeans
os.chdir('D:/Term-2/MachineLearning/26-02-2019(clustering)')
```

```python
tnr = plt.imread('TNR.jpg')
```

```python
plt.imshow(tnr)
```

```python
x,y,z = tnr.shape
```

```python
img1 = tnr.reshape(x*y,z)
```

```python
img1.shape
```

```python
mod = KMeans(n_clusters = 3)
modfit = mod.fit(img1)
```

```python
cent = modfit.cluster_centers_
cent
```

```python
lab = modfit.labels_
lab
```

```python
np.bincount(lab)
```
