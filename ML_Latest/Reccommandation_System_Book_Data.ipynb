```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import sklearn
```

```python
book_data = pd.read_csv('D:/Term_2/ML/ML_Sample_Data_Sets/book_recom.csv')
book_data.head()
```

```python
data_pivot = book_data.pivot(index = 'bookTitle', columns = 'userID',values = 'bookRating').fillna(0)
data_pivot
```

```python
book_data_matrix = csr_matrix(data_pivot.values)
book_data_matrix

# original values in matrix 198x11337
# in o/p 11110 is non zero values in sparse_matrix
```

```python
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute' ) # brute force linear search algorithm

model_knn.fit(book_data_matrix)
```

# Recommendation for reading:
    
  Find the 5 similar (with maximum similarity - means lowest distance) books as that ot the first book '1st to Die: A novel'.
    
  The function kneighbors() is used to find k number of neighbors of a point.
    
  we need to compute 6 neighbors from book1 to all other books. Note that the shortest distance possible now is 0. which is  book1 to book1 itself. Hence we need to take 6 neighbors, instead 5

```python
query_index1 = 0 # 1st book
distances,indices = model_knn.kneighbors(
data_pivot.iloc[query_index1,:].values.reshape(1,-1), # quary_index like user input i.e bookname, : means all columns
n_neighbors = 6)
```

```python
distances # 0 is combination with that itself, other values least distances
```

```python
print("Recommendations for {0}\n" .format(data_pivot.index[query_index1]))
for i in indices.flatten()[1:]:
    print(data_pivot.index[i])
```
