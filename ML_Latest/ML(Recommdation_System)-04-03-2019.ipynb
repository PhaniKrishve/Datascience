```python
book=pd.read_csv("book_recom.csv")
book.head()
```

```python
book1 =book[['bookTitle','userID','bookRating']]
book1.head()
```

```python
book_pivot = (book1.pivot_table(columns='userID',index='bookTitle',values='bookRating').fillna(0))
```

```python
book_pivot.head()
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
?NearestNeighbors
```

```python
book_matrix = csr_matrix(book_pivot.values)
book_matrix
```

```python
model_knn =NearestNeighbors(metric = 'cosine',algorithm = 'brute')
```

```python
model_knn.fit(book_matrix)
```

```python
model_knn.fit(book_pivot)
```

```python
query_index1 = 0
```

```python
distance,indicies = model_knn.kneighbors(book_pivot.iloc[query_index1,:].reshape(1,-1),n_neighbors=6)
```

```python
pro =pd.read_csv("products - Sheet1.csv")
```

```python
index=pro['ProductId']
title = pro['Title']
```

```python
pro1 = pro.drop(['Title','Image URL','Division','Color','ProductId'],axis=1)
```

```python
pro_dum=pd.get_dummies(pro1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vector =CountVectorizer(stop_words='english',min_df=5)
```

```python
count_vector.fit(title)
dtm = count_vector.transform(title)
```

```python
df_dtm = pd.DataFrame(dtm.toarray(),columns = count_vector.get_feature_names())
df_dtm.head()
```

```python
dtm1 = dtm.toarray()
dtm2  = pd.DataFrame(dtm1)
```

```python
product =pd.concat([pro_dum,dtm2],axis=1)
```

```python
product.head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model_kn = NearestNeighbors(metric='cosine',algorithm='brute')
model_kn.fit(product)
```

```python
test = product.iloc[0,:].reshape(1,-1)
```

```python
distances,indices = model_kn.kneighbors(test,n_neighbors=10)
ind = indicies[0]
recom = index[indices[0]]
```
