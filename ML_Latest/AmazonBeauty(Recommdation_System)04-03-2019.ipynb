```python
beauty  = pd.read_json("Beauty_5.json",lines=True)
```

```python
beauty.head()
```

```python
ratings=pd.read_csv("ratings_Beauty.csv",header=None)
ratings.columns= ['reviewerID','asin','rating','unixReviewTime']
ratings=ratings.drop(['unixReviewTime'],axis=1)
ratings.head()
```

```python
ratings[ratings.reviewerID=='A1YJEY40YUW4SE']
```

```python
beauty_rating = pd.merge(beauty,ratings,on=['asin','reviewerID'])
```

```python
beauty_rating.head()
```

```python
beauty_pivot = beauty_rating.pivot("asin",'reviewerID','rating').fillna(0)
```

```python
beauty_pivot.head()
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
beauty_matrix = csr_matrix(beauty_pivot.values)
beauty_matrix
```

```python
model_knn =NearestNeighbors(metric = 'cosine',algorithm = 'brute')
```

```python
model_knn.fit(beauty_matrix)
```

```python
distance,indicies = model_knn.kneighbors(beauty_pivot.iloc[0,:].reshape(1,-1),n_neighbors=6)
```
