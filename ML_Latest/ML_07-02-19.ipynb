```python
import pandas as pd
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("D:/Term-2/MachineLearning/nursery.data.txt",sep=',')
```

```python
data.columns = ['parents','has_nurs','form','children','housing','finance','social','health','tar']
```

```python
data
```

```python
data.isnull().sum()
```

```python
y=data['tar']
x = data.drop(['tar'],axis=1)
x1 = pd.get_dummies(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score,auc
```

```python
mod = BernoulliNB()
mod.fit(x_train,y_train)
```

```python
y_pred = mod.predict(x_test)
y_pred
```

```python
ac = accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred1,y_test)
```

```python
mod2 = GaussianNB()
mod2.fit(x_train,y_train)
```

```python
y_pred2 = mod2.predict(x_test)
y_pred2
```

```python
accuracy_score(y_pred2,y_test)
```

# HR Analytics

```python
hr = pd.read_csv("D:/Term-2/MachineLearning/HR Analytics.csv")
hr
```

```python
y=hr['Attrition']
#y=y.drop(['id'],axis=1)
x = hr.drop(['Attrition'],axis=1)
x1 = pd.get_dummies(x)
```

```python
x_train,x_test,y_train,y_test  = train_test_split(x1,y,test_size =0.3,random_state=42)
```

```python
test_sam = x_test.iloc[0]
test_sam
```

```python
from scipy.spatial.distance import euclidean
```

```python
dist_sort  = pd.DataFrame()
for i,row in x_train.iterrows():
    dist =euclidean(row.values,test_sam.values)
    dist_sort=dist_sort.append({'ind':i,'val':dist},ignore_index = True)
    
dist_so = dist_sort.sort_values(by="val")
dist_so.head(5)
```

```python
hr.iloc[[1385,817,1125,1007,760]]
```

```python
model = KNeighborsClassifier(n_neighbors=20,algorithm='auto')
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
p={'n_neighbors':[1,3,5,7,9],'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute']}
```

```python
knn_grid = KNeighborsClassifier()
modd = GridSearchCV(knn_grid,param_grid=p,cv=5)
```

```python
modd.fit(x_train,y_train)
print("fgdf",modd.best_params_)
```

```python
y_score = modd.predict_proba(x_test)
```

```python
fpr,tpr,th=roc_curve(y_test,y_score[:,1],pos_label=1)
AUC=auc(fpr,tpr)
print(AUC)
plt.plot(fpr,tpr)
```

```python
accuracy_score(y,y_pred)
```
