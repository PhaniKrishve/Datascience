```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from scipy.spatial.distance import euclidean
from sklearn.model_selection import GridSearchCV
```

```python
data1 = pd.read_csv('products - Sheet1.csv')
data1.head()
```

```python
data1 = data1.set_index('ProductId')
data1.head()
```

```python
data2 = data1.drop(['Title','Image URL'],axis=1)
data2.head()
```

```python
data2.isnull().sum()
```

```python
data2.dropna()
```

```python
y = data2[['Division','Category']]
x = data2.drop(['Division','Category'],axis=1)
```

```python
data3 = pd.get_dummies(x)
data3.head()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(data3,y,test_size = 0.25,random_state = 42)
```

```python
data3.columns
```

```python
test_sample = xtest.iloc[0]
dist_mat = pd.DataFrame(columns = ['Index','distance'])
for Index,row in xtrain.iterrows():
    dist = euclidean(test_sample.values,row.values)
    dist_mat = dist_mat.append({'Index':Index,'distance':dist},ignore_index=True)
    
dist_sort = dist_mat.sort_values(by='distance')
knn_sore = dist_sort
knn_sore
```
