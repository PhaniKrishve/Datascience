```python
top_movies['popularity_index']=top_movies['No_of_reviewers']*top_movies['IMD_Rating']
```

```python
#Extracting a new feature called generes from the genre1,genre2,genre3
top_movies['genre1'] = np.where(top_movies.genre1.isnull(),"",top_movies['genre1'])
top_movies['genre2'] = np.where(top_movies.genre2.isnull(),"",top_movies['genre2'])
top_movies['genre3'] = np.where(top_movies.genre3.isnull(),"",top_movies['genre3'])
top_movies['genres']=top_movies['genre1']+top_movies['genre2']+top_movies['genre3']
top_movies.genres.value_counts()
```

```python
#Apply count_vectorizator to keywords list column
keywords =top_movies['keywords_list']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector =CountVectorizer(stop_words='english',min_df=5)
count_vector.fit(keywords)
dtm = count_vector.transform(keywords)
df_dtm = pd.DataFrame(dtm.toarray(),columns = count_vector.get_feature_names())
df_dtm.head()
```

```python
index=top_movies['movie_name']
movies=top_movies[['genres','star1','star2','star3','popularity_index','writer1']]
movies.head()
```

```python
movies_all  = pd.concat([movies,df_dtm],axis=1)
movies_all.head()
```

# Building a Recommdation system for 250 movies 

```python
movie_dummies=pd.get_dummies(movies_all)
movie_dummies.head()
```

```python
from sklearn.neighbors import NearestNeighbors
model_kn = NearestNeighbors(metric='cosine',algorithm='brute')
model_kn.fit(movie_dummies)
```
