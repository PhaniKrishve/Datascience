```python
import pandas as pd
import numpy as np
from ipywidgets import widgets,interact
from matplotlib import pyplot as plt
import seaborn as sns
import os
os.chdir("D:/Term-2/Data Visualization")
```

```python
un_data = pd.read_excel("UN_Data.xlsx")
un_data.head()
```

```python
un_data.shape
```

```python
un_data1 = un_data[un_data.Month!='Total']
un_data1.drop_duplicates(inplace = True)
un_data1.shape
```

```python
country = widgets.Dropdown(options = ['All']+list(un_data1['Country'].unique()),
                        value = 'All',
                        description = "Select the country")
```

```python
un_data2 = un_data1[(un_data1.Year == 2009) & (un_data1.Country=='Italy')]
un_data2
```

```python
def plotit(x):
    if (country.value=='All') :
        df2 = un_data1.groupby(['Year','Month'])['Value'].sum().reset_index()
    else:
        df2 = un_data1[un_data1.Country==country.value]
    df2_wide =df2.pivot('Month','Year','Value')
    #index=['January']
    plt.rcParams['figure.figsize']=[15,5]
    sns.heatmap(df2_wide,linewidths=0.5,cmap = 'hot_r')
    plt.show()
interact(plotit,x=country)
```
