```python
import pandas as pd
import numpy as np
import os 
#os.chdir('c://users//administrator//desktop/NYC_cabs/')

import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from folium.plugins import HeatMapWithTime
from folium.plugins import HeatMap
```

```python
y = pd.read_csv('train.csv')
y.head()
```

```python
y.shape
```

```python
yy_cabs_sample = y.head(100000)
```

```python
yy_cabs_sample.pickup_datetime = pd.to_datetime(yy_cabs_sample.pickup_datetime, format='%Y-%m-%d %H:%M:%S')
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
yy_cabs_sample['month'] = yy_cabs_sample.pickup_datetime.apply(lambda x: x.month)
yy_cabs_sample['week'] = yy_cabs_sample.pickup_datetime.apply(lambda x: x.week)
yy_cabs_sample['day'] = yy_cabs_sample.pickup_datetime.apply(lambda x: x.day)
yy_cabs_sample['hour'] = yy_cabs_sample.pickup_datetime.apply(lambda x: x.hour)
```

```python
## ADDING NEW COLUMN TRIP COUNT

yy_cabs_sample['Trip'] = 1
```

```python
yy_cabs_sample['pickup_longitude_round']=np.round(yy_cabs_sample['pickup_longitude'],2)
yy_cabs_sample['pickup_latitude_round']=np.round(yy_cabs_sample['pickup_latitude'],2)
yy_cabs_sample['dropoff_longitude_round']=np.round(yy_cabs_sample['dropoff_longitude'],2)
yy_cabs_sample['dropoff_latitude_round']=np.round(yy_cabs_sample['dropoff_latitude'],2)
```

```python
def generateBaseMap(default_location=[40.693943, -73.985880], default_zoom_start=11):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
base_map = generateBaseMap()
base_map
```

```python
from folium.plugins import HeatMap
base_map = generateBaseMap()
HeatMap(data=yy_cabs_sample[['pickup_latitude_round', 'pickup_longitude_round', 'Trip']].groupby(['pickup_latitude_round', 'pickup_longitude_round']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
base_map
```

```python
hours = []
for hour in yy_cabs_sample.hour.sort_values().unique():
    hours.append(yy_cabs_sample.loc[yy_cabs_sample.hour == hour, ['pickup_latitude_round', 'pickup_longitude_round', 'Trip']].groupby(['pickup_latitude_round', 'pickup_longitude_round']).sum().reset_index().values.tolist())
```

```python
base_map = generateBaseMap()
base_map = generateBaseMap(default_zoom_start=10)
HeatMapWithTime(hours, radius=7, gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
base_map
```
