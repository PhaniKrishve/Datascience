```python
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
#converting the load data into a wide data
flights_wide = flights.pivot("month","year",'passengers')
flights_wide
```

```python
#ploting heat map
plt.rcParams['figure.figsize']=[10,8]

#plot the heat map
plt.imshow(flights_wide,cmap="hot")
#Bring year on x-axis
plt.xticks(range(len(flights_wide.columns)),flights_wide.columns)
plt.yticks(range(len(flights_wide.index)),flights_wide.index)


plt.colorbar()


#
plt.show()
```

```python
#plotting  the same heat map in seaborn
sns.heatmap(flights_wide,linewidths=0.5,cmap = "hot_r",annot=True,fmt ='d')
```
